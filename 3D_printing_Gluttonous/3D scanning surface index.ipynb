```python
import trimesh
import os

# Define your folder path
folder_path = r"C:\Users\qiong\Desktop\Manuscript\Manuscript M2\Method exlporation"

# Verify folder exists
if not os.path.isdir(folder_path):
    print(f"‚ùå Error: '{folder_path}' is not a valid directory.")
else:
    mfiles = os.listdir(folder_path)
    print(f"üìÇ Found {len(mfiles)} files in folder.\n")

    for filename in mfiles:
        file_path = os.path.join(folder_path, filename)

        # Skip non-PLY files
        if not filename.lower().endswith('.ply'):
            continue

        try:
            # Load the mesh
            mesh = trimesh.load_mesh(file_path, force='mesh')

            # Ensure it's a valid mesh
            if not isinstance(mesh, trimesh.Trimesh):
                print(f"‚ö†Ô∏è Skipping {filename}: Not a valid mesh.")
                continue

            # Compute surface area
            surface_area = mesh.area
            print(f"{filename}\tSurface Area: {surface_area:.4f} square units")

        except Exception as e:
            print(f"‚ùó Error processing {filename}: {e}")
```

```python
import trimesh
import os

# Define your folder path
folder_path = r"C:\Users\qiong\Desktop\test"

if not os.path.isdir(folder_path):
    print(f"‚ùå Error: '{folder_path}' is not a valid directory.")
else:
    mfiles = os.listdir(folder_path)
    print(f"üìÇ Found {len(mfiles)} files in folder.\n")

    for filename in mfiles:
        file_path = os.path.join(folder_path, filename)

        # Process only common 3D formats
        if not filename.lower().endswith(('.ply', '.obj', '.stl')):
            continue

        try:
            mesh = trimesh.load(file_path)

            # Convert Scene ‚Üí Trimesh if necessary
            if isinstance(mesh, trimesh.Scene):
                mesh = trimesh.util.concatenate(mesh.dump())

            if not isinstance(mesh, trimesh.Trimesh):
                print(f"‚ö†Ô∏è Skipping {filename}: Not a valid mesh.")
                continue

            surface_area = mesh.area
            print(f"{filename}\tSurface Area: {surface_area:.4f} square units")

        except Exception as e:
            print(f"‚ùó Error processing {filename}: {e}")
```
